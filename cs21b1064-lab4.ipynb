```python
#1)
def push(stack, element):
    stack.append(element)
    print("Element {} pushed to stack".format(element))

def pop(stack):
    if len(stack) == 0:
        print("Stack is empty")
    else:
        print("Element {} popped from stack".format(stack.pop()))
        
def peek(stack):
    if len(stack) == 0:
        print("Stack is empty")
    else:
        print("Top element of stack is {}".format(stack[-1]))
    
def print_stack(stack):
    if len(stack) == 0:
        print("Stack is empty")
    else:
        print("Stack elements are: ", end = "")
        for i in range(len(stack)-1, -1, -1):
            print(stack[i], end = " ")
        print()
        

stack = []
while True:
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Print")
    print("5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        element = int(input("Enter element to push: "))
        push(stack, element)
    elif choice == 2:
        pop(stack)
    elif choice == 3:
        peek(stack)
    elif choice == 4:
        print_stack(stack)
    elif choice == 5:
        break
    else:
        print("Invalid choice")
        print()

            



```

```python
#2)
def enqueue(queue, element):
    queue.append(element)
    print("Element {} enqueued to queue".format(element))

def dequeue(queue):
    if len(queue) == 0:
        print("Queue is empty")
    else:
        print("Element {} dequeued from queue".format(queue.pop(0)))
        
def peek(queue):
    if len(queue) == 0:
        print("Queue is empty")
    else:
        print("Front element of queue is {}".format(queue[0]))
        
def print_queue(queue):
    if len(queue) == 0:
        print("Queue is empty")
    else:
        print("Queue elements are: ", end = "")
        for i in range(len(queue)):
            print(queue[i], end = " ")
        print()
        
queue=[]
while True:
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Peek")
    print("4. Print")
    print("5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        element = int(input("Enter element to enqueue: "))
        enqueue(queue, element)
    elif choice == 2:
        dequeue(queue)
    elif choice == 3:
        peek(queue)
    elif choice == 4:
        print_queue(queue)
    elif choice == 5:
        break
    else:
        print("Invalid choice")
        print()
```
